```python
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
import os
os.chdir("D:/Training/Academy/ML_Python/Datasets/")
result = adfuller(df['IPG2211A2N'], maxlag=10)
print("P-Value =", result[1])
if result[1] < 0.05:
    print("Time Series is Stationary")
else:
    print("Time Series is not Stationary")

diff1 = df['IPG2211A2N'].diff().iloc[1:]
result = adfuller(diff1, maxlag=10)
print("P-Value =", result[1])
if result[1] < 0.05:
    print("Time Series is Stationary")
else:
    print("Time Series is not Stationary")

plot_acf(df['IPG2211A2N'], lags=30,alpha=None)
plt.show()

diff1 = df['IPG2211A2N'].diff().iloc[1:]
plot_acf(diff1, lags=30,alpha=None)
plt.show()

plot_pacf(diff1, lags=30,alpha=None)
plt.show()
```

```python
#arima
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import root_mean_squared_error
from statsmodels.tsa.stattools import adfuller
import os
os.chdir("D:/Training/Academy/ML_Python/Datasets/")
fig, ax = plt.subplots(2,1, figsize=(12,8))
plot_acf(diff1, ax=ax[0])
plot_pacf(diff1, ax=ax[1])
plt.show()


y_train = df['Milk'].iloc[:-12]
y_test = df['Milk'].iloc[-12:]
model = ARIMA(y_train,order=(1,1,2))
model_fit = model.fit()
print('Coefficients: %s' % model_fit.params)

predictions = model_fit.predict(start=len(y_train), end=len(y_train)+len(y_test)-1,dynamic=False)
root_mean_squared_error(y_test, predictions)

y_train.plot(color="blue", label='Train')
y_test.plot(color="pink", label='Test')
rmse = np.round( root_mean_squared_error(y_test, predictions), 2)
plt.text(100, 600, f"RMSE = {rmse}")
predictions.plot(color="purple", label='Forecast')
plt.legend(loc='best')
plt.show()
```

```python
#auto-arima
from pmdarima.arima import auto_arima
model = auto_arima(y_train, trace=True, error_action='ignore',  suppress_warnings=True)
y_pred = model.predict(n_periods=len(y_test))
forecast = pd.DataFrame(y_pred, index = y_test.index, columns=['Prediction'])
plt.plot(y_train, label='Train',color="blue")
plt.plot(y_test, label='Valid',color="pink")
rmse = np.round( root_mean_squared_error(y_test, y_pred), 2)
plt.text(100, 600, f"RMSE = {rmse}")
plt.plot(forecast, label='Prediction',color="purple")
plt.show()
```

```python
#seasonal arima, sarima
s_model = auto_arima(y_train, trace=True, error_action='ignore',
                   suppress_warnings=True,seasonal=True,m=12)

y_pred = s_model.predict(n_periods=len(y_test))
forecast = pd.DataFrame(y_pred,index = y_test.index, columns=['Prediction'])
plt.plot(y_train, label='Train',color="blue")
plt.plot(y_test, label='Valid',color="pink")
rmse = np.round( root_mean_squared_error(y_test, y_pred), 2)
plt.text(100, 600, f"RMSE = {rmse}")
plt.plot(forecast, label='Prediction',color="purple")
plt.show()
```
